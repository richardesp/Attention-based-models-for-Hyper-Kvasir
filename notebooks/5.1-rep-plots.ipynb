```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset_path = '../data/hyper-kvasir-dataset-vanilla'
```

```python
# Count the number of images in the final folder
folders = os.listdir(dataset_path)

images = []

for folder in folders:
    current_path = os.path.join(dataset_path, folder)

    if os.path.isdir(current_path):
        images.append((os.listdir(current_path)[3], folder, len(os.listdir(current_path))))
```

```python
# Get an array of the second element of each tuple

titles = ['Polyps', 'Cecum', 'Ileum', 'Hemorrhoids', 'Z-line', 'Pylorus']

img_arr = []

for image in images:

    if image[1] in [title.lower() for title in titles]:
        img_arr.append(plt.imread(os.path.join(dataset_path, image[1], image[0])))

print(img_arr.__len__())
```

```python
def display_images(images, titles, rows=2, cols=3):
    fig, axes = plt.subplots(rows, cols, figsize=(cols * 2, rows * 2))

    # Primera fila con 12 imgenes
    for i, ax in enumerate(axes[0, :]):
        ax.imshow(images[i])
        ax.set_title(titles[i])
        ax.axis('off')

    # Segunda fila con 11 imgenes, centradas
    start_idx = cols
    end_idx = start_idx + 11
    offset = (cols - 11) // 2
    for i, ax in enumerate(axes[1, offset:end_idx+offset]):
        ax.imshow(images[start_idx + i])
        ax.set_title(titles[start_idx + i])
        ax.axis('off')

    # Esconder los ejes de los espacios vacos en la segunda fila
    for ax in axes[1, :offset]:
        ax.axis('off')
    for ax in axes[1, end_idx+offset:]:
        ax.axis('off')

    plt.tight_layout()
    plt.show()

display_images(img_arr, titles)
```
